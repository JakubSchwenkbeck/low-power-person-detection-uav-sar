```python
import matplotlib.pyplot as plt

def plot_benchmarks(items, save=None):
    names = [x.get("model", "") for x in items]
    energy = [x.get("energy", 0) for x in items]
    fps = [x.get("fps", 0) for x in items]
    map_ = [x.get("map", 0) for x in items]
    iou = [x.get("iou", 0) for x in items]

    fig, ax = plt.subplots(2, 2, figsize=(10, 7))
    x = range(len(names))

    ax[0, 0].bar(x, energy); ax[0, 0].set_title("Energy"); ax[0, 0].set_ylabel("J")
    ax[0, 1].bar(x, fps);    ax[0, 1].set_title("FPS");    ax[0, 1].set_ylabel("frames/s")
    ax[1, 0].bar(x, map_);   ax[1, 0].set_title("mAP");    ax[1, 0].set_ylabel("score")
    ax[1, 1].bar(x, iou);    ax[1, 1].set_title("IoU");    ax[1, 1].set_ylabel("score")

    for a in ax.flat:
        a.set_xticks(list(x))
        a.set_xticklabels(names, rotation=30, ha="right")

    fig.suptitle("Benchmarks")
    fig.tight_layout()

    if save:
        plt.savefig(save, dpi=200, bbox_inches="tight")
    else:
        plt.show()
        
benchmarks = [
    {"model": "efficientdet_d0_size_float16", "energy": 1.8, "fps": 12.3, "map": 0.42, "iou": 0.55},
    {"model": "yolo11n_latency_dynamic", "energy": 1.2, "fps": 18.7, "map": 0.38, "iou": 0.51},
]

plot_benchmarks(benchmarks) 
```
