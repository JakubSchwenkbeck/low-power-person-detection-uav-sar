```python
import matplotlib.pyplot as plt

def plot_benchmarks(benchmarks: list[dict], output_path: str | None = None):
    models = [b.get("model", "") for b in benchmarks]
    energy = [b.get("energy", 0) for b in benchmarks]
    fps = [b.get("fps", 0) for b in benchmarks]
    map_ = [b.get("map", 0) for b in benchmarks]
    iou = [b.get("iou", 0) for b in benchmarks]

    fig, axs = plt.subplots(2, 2, figsize=(12, 8))
    x = range(len(models))

    axs[0, 0].bar(x, energy)
    axs[0, 0].set_title("Energy")
    axs[0, 0].set_ylabel("J")
    axs[0, 0].grid(True, axis="y", alpha=0.3)

    axs[0, 1].bar(x, fps, color="#4c78a8")
    axs[0, 1].set_title("FPS")
    axs[0, 1].set_ylabel("frames/s")
    axs[0, 1].grid(True, axis="y", alpha=0.3)

    axs[1, 0].bar(x, map_, color="#f58518")
    axs[1, 0].set_title("mAP")
    axs[1, 0].set_ylabel("score")
    axs[1, 0].grid(True, axis="y", alpha=0.3)

    axs[1, 1].bar(x, iou, color="#54a24b")
    axs[1, 1].set_title("IoU")
    axs[1, 1].set_ylabel("score")
    axs[1, 1].grid(True, axis="y", alpha=0.3)

    for ax in axs.flat:
        ax.set_xticks(list(x))
        ax.set_xticklabels(models, rotation=30, ha="right")

    fig.suptitle("Benchmark Comparison", y=0.98)
    fig.tight_layout()

    if output_path:
        plt.savefig(output_path, dpi=200, bbox_inches="tight")
    else:
        plt.show()

        
benchmarks = [
    {"model": "efficientdet_d0_size_float16", "energy": 1.8, "fps": 12.3, "map": 0.42, "iou": 0.55},
    {"model": "yolo11n_latency_dynamic", "energy": 1.2, "fps": 18.7, "map": 0.38, "iou": 0.51},
]

plot_benchmarks(benchmarks) 
```
